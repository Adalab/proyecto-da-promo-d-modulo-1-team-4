```python
import random    #para seleccionar opciones aleatorias para la maquina entre piedra papel o tiejra
import emoji      #para agregar emojis en los prints

# Variables previas
registro_jugadores = {}  #diccionario para almacenar los jugadores y sus estadísticas (partinas ganadas, perdinas, empatadas)
opciones_respuesta = ["piedra", "papel", "tijera"]    #lista que tiene las unicas opciones del juego: piedra apel o tijera

# Función para registrar a un participante. sie l jugador ya esta registrado en registro_jugadores muestra un mensaje y si no esta añadido, lo añade con las estadicticas de de cero (partidas, ganas, jugas y empatadas) y le muestra un mensaje.
def vamos_participante(nombre, registro):
    if nombre in registro:
        print(f"\n¡Bienvenido {nombre}! {emoji.emojize(':grinning_face:')}")
    else:
        registro[nombre] = {"ganadas": 0, "perdidas": 0, "empatadas": 0}
        print(f"\n¡Hola, {nombre}! Bienvenido al juego. {emoji.emojize(':grinning_face:')}")

def iniciar_juego():
    # Inicio del programa con gestion de errores, da la bienvenida, pide el nombre al jugador y con strip elimina los espacios 
    try:
        print("¡Bienvenido al juego de Piedra, Papel o Tijera!")
        nombre_usuario = input("Introduce tu nombre: ").strip()

        # Registramos al nuevo usuario llama a la funcion
        vamos_participante(nombre_usuario, registro_jugadores)

        # Mensaje inicial
        print(f"\n¡Hola! {nombre_usuario} \U0001F600")
        print("Hasta ahora eras tú quien cumplía los sueños de tus viajeros. Ahora nos toca a nosotros cumplir el viaje de tus sueños.")
        print("Piedra, papel, tijera... ¡y vuela!")
        print("¡Vamos a volar! ✈️")
        print("---------------------------")

        # Variables del juego
        partidas_jugadas = 0     #esta es el contdor de partidas
        partidas_ganadas_usuario = 0   #el contador de las partidas del usuario
        partidas_ganadas_maquina = 0    #el contador de las maquina
        partidas_empatadas = 0    #el contador de las partidas empatas
        total_partidas = 2     #el paximo de las partidas pemitidas para cada jugador


        # Ciclo del juego propiamente dicho
        while partidas_jugadas <= total_partidas:    #El bucle se ejecuta mientras partidas_jugadas sea menor que total_partidas
            respuesta_usuario = input("Elige piedra, papel o tijera. Introduce 'salir' para terminar el juego: ").lower().strip()
            if respuesta_usuario == "salir":
                print("¡Gracias por jugar! Saliendo del juego...")
                break
            elif respuesta_usuario not in opciones_respuesta:   #Si el usuario escribe algo que no sea "piedra", "papel" o "tijera", se muestra un mensaje de error y vuelve a pedir una entrada válida.
                print("\nOpción incorrecta. Elige una de las opciones disponibles.")
                continue
                #logica del juego pura y dura la maquina selecciona una respuesta aleatoria y despues se va incrementando el numero de partidas jugadas para llegar el conteo en las estadisticas de cada jugador
            respuesta_maquina = random.choice(opciones_respuesta)
            partidas_jugadas += 1
            print(f"\nEs la partida número {partidas_jugadas}")
            print(f"Has elegido: {respuesta_usuario}. El oponente ha escogido: {respuesta_maquina}.")

            if respuesta_usuario == respuesta_maquina:
                partidas_empatadas += 1
                print("---------------------------")
                print(f"\n¡Empate! Lleváis {partidas_empatadas} partidas empatadas.")
                print("---------------------------")
            elif (respuesta_usuario == "piedra" and respuesta_maquina == "tijera") or \
                (respuesta_usuario == "papel" and respuesta_maquina == "piedra") or \
                (respuesta_usuario == "tijera" and respuesta_maquina == "papel"):
                partidas_ganadas_usuario += 1
                print("---------------------------")
                print(f"\n¡Has ganado! Llevas {partidas_ganadas_usuario} partidas ganadas.")
                print("---------------------------")
            else:
                partidas_ganadas_maquina += 1
                print("---------------------------")
                print(f"\nHas perdido. El ordenador ha ganado {partidas_ganadas_maquina} partidas.")
                print("---------------------------")

        # Actualizar el registro del jugador con sus estadisticas
        registro_jugadores[nombre_usuario]["ganadas"] += partidas_ganadas_usuario
        registro_jugadores[nombre_usuario]["perdidas"] += partidas_ganadas_maquina
        registro_jugadores[nombre_usuario]["empatadas"] += partidas_empatadas

        # Mensaje final
        print("\n--- RESULTADO FINAL ---")
        print(f"Partidas jugadas: {partidas_jugadas}")
        print(f"Ganadas por ti: {partidas_ganadas_usuario}")
        print(f"Ganadas por la máquina: {partidas_ganadas_maquina}")
        print(f"Empates: {partidas_empatadas}")
        print("¡ALEA JACTA EST!")

        # Mostrar estadísticas del usuario
        print("\n--- ESTADÍSTICAS ACTUALES ---")
        print(f"Estadísticas de {nombre_usuario}:")
        print(f"Ganadas: {registro_jugadores[nombre_usuario]['ganadas']}")
        print(f"Perdidas: {registro_jugadores[nombre_usuario]['perdidas']}")
        print(f"Empatadas: {registro_jugadores[nombre_usuario]['empatadas']}")

    except Exception as e:
        print("\n¡Ha ocurrido un error inesperado!")
        print(f"Detalles del error: {e}")
```

```python
iniciar_juego()
```
